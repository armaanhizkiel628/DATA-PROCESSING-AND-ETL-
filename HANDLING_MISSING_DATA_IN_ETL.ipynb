# Q1. What are the most common reasons for missing data in ETL pipelines?

answer1

1. Some of the most common reasons for missing data in ETL pipelines are:
* Source system issues: Sometimes the original data source does not collect certain data, or the system is down, so the data is missing.

* Data extraction errors: During extraction, data can be lost because of connection problems, timeouts, or incorrect queries.

* Transformation problems: Errors in transformation logic, such as wrong filters, joins, or data type conversions, can cause some records or fields to be dropped.

* Data quality issues: Missing values can already exist in the source data due to human error, incomplete forms, or optional fields.

* Schema changes: If the source schema changes (like a column is renamed or removed) and the ETL pipeline is not updated, data may not load correctly.

# Q2. Why is blindly deleting rows with missing values considered a bad practice in ETL?

answer2

1. Blindly deleting rows with missing values is considered a bad practice in ETL because it can cause important data to be lost.

2. It can also introduce bias, since the remaining data may no longer represent the real situation accurately.

3. deleting rows without understanding why the data is missing can lead to incorrect analysis and poor decision-making


# Q3. Explain the difference between:Listwise deletion Column deletion Also mention one scenario where each is appropriate.

answer3

1. listwise deletion = Listwise deletion means removing an entire row if it has at least one missing value in any column

* Appropriate scenario: This method is suitable when the dataset is large and only a very small number of rows have missing values, so deleting them does not significantly affect the overall analysis.

2. column deletion = Column deletion means removing an entire column if it contains many missing values
  

* Appropriate scenario: This is appropriate when a column has a very high percentage of missing data and is not important for analysis, such as an optional field that is rarely filled in.  

# Q4. Why is median imputation preferred over mean imputation for skewed data such as income?

1. Median imputation is better than mean imputation for skewed data like income because the mean is easily affected by extreme values, such as very high salaries. These extreme values can change the average and make it inaccurate for most people. The median, on the other hand, is not affected much by outliers and represents the middle value of the data. Because of this, using the median gives a more realistic value for missing income data.


# Q5. What is forward fill and in what type of dataset is it most useful?

answer 5

1. Forward fill is a method used to handle missing data where a missing value is filled with the previous available value in the dataset. It means the last known value is carried forward to replace the missing one.

2. This method is most useful in time-series data, such as stock prices, sensor readings, or daily temperature data, where values change gradually over time and the previous value is a reasonable estimate for the missing one.


# Q6. Why should flagging missing values be done before imputation in an ETL workflow?


answer

1. Flagging missing values before imputation is important because it helps keep track of which data was originally missing. Once values are filled in, it becomes hard to tell which ones were real and which ones were added later. By flagging them first, analysts can study patterns in missing data and understand if it affects results. It also helps with transparency and makes the ETL process easier to audit and debug later.
